Uploading the file "pedestrians"

```python
import pandas as pd
df = pd.read_csv("Pedestrians.csv")
```

Correcting wrong columns 
"Date_Time" to datetime
"Hourly_Conts" to int after removing comma

```python
df['Date_Time'] = pd.to_datetime(df['Date_Time'])

#removing comma and transforming to integer
df['Hourly_Counts'] = df['Hourly_Counts'].str.replace(',', '').astype(int)
```

Setting the datetime column as index


```python
df = df.set_index('Date_Time')
```

Sorting the index to get the right order

```python
df= df.sort_index(axis = 0, ascending = True)
```
