Checking for duplicates

delete not needed columns, since Sensor is the same troughout it can be deleted

Sorting the index to get the right order

```python
df2= df2.sort_index(axis = 0, ascending = True)
df2
```

Ploting the time series

```python
import seaborn as sns
sns.lineplot(x=df2.index, y='Hourly_Counts', data = df2)
```

```python
#the past errors can be used to predict future patterns
#maybe there is a pattern in errors
#bootstrapping or monte-carlo will be used

##how does it work:
#we predict a value and take a random error from the past and add/substract it from the predicted value, we repeat that thhousand of time and hence calculate our confidence interval 
#for the prediciton
```

```python
df2.head()
```

```python
df3 = df2[['Month', 'Day', 'Time', 'Hourly_Counts']]
```

```python
df3
```

```python
df3=df2.loc[:'2019-12-31',:]
```

```python
df3
```

```python
df3['Trend'] = range(1,len(df3)+1)
```

```python
import statsmodels.formula.api as sm
model = sm.ols(formula="Hourly_Counts ~ Trend + Month + Time + Day", data=df3)
results = model.fit()
results.summary()
```

```python
results.predict(df3)
```

```python
df3['Counts_OOS']=results.predict(df3)
```

```python
df3
```

```python
dfplot = df3.filter(['Hourly_Counts', 'Counts_OOS'])
sns.lineplot(data=dfplot)
```
