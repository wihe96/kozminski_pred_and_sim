Uploading the file "pedestrians"

```python
import pandas as pd
import seaborn as sns
import numpy as np


df = pd.read_csv("Pedestrians.csv", thousands = ',') #thousands operator makes the count integer
```

```python
df.head(5)
```

Checking df for right formats


```python
print(df.dtypes)
df.head(5)
```

Correcting wrong columns 
"Date_Time" to datetime


```python
df.index= df.Date_Time.astype('datetime64')
```

```python
df.head(5)
```

Checking for duplicates

Make a decision on what to do with them:



```python
duplicate = df[df.duplicated(subset = 'Date_Time', keep = 'first')]
duplicate

#or df[df.Date_Time.duplicated() == True]
```

Check the duplicates

it comes all from the same year and time (midnight)

```python
duplicate.describe()
```

```python
df2 = df.drop_duplicates(subset ='Date_Time', keep = False)
```

Do mean count for midnight for whole sample

```python
df2[df2['Time']==0.0].mean()
```

delete not needed columns, since Sensor is the same troughout it can be deleted

```python
del df2['Sensor_ID']
del df2['Sensor_Name']
del df2['ID']
del df2['Date_Time']
```

Sorting the index to get the right order

```python
df2= df2.sort_index(axis = 0, ascending = True)
df2
```

Ploting the time series

```python
import seaborn as sns
sns.lineplot(x=df2.index, y='Hourly_Counts', data = df2)
```
